```python
#Banking Application

class Atm:

  # constructer -----> a function inside the class  ----> jb bhi variables bnaoge to constructer ke andar bnaoge


  #constructer (special function) ----> superpower ---> iss function ko explicitly call krne ki zrurt nhi hai   ---> iska code automatically trigger ho jaata hai jb uss class ka object bnate hai

  def __init__(self):    # ---> rule 1 variables constructer ke andar bnaoge
    #print(id(self))
    self.pin=''          #------> rule 2 : variable name ke aage self lgaoge
    self.balance=0
    self.menu()         # constructer ke andar hai to auromatically print ho gya

  def menu(self):
    user_input=input('''
    Hi how can I help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    ''')
    if user_input == '1':
      #to create pin
      self.create_pin()

    elif user_input == '2':
      #to change the pin
      self.change_pin()

    elif user_input == '3':
      #to check balance
      self.check_balance()

    elif user_input == '4':
      #to withdraw
      self.withdraw()

    else:
      #to exit
      exit()

  def create_pin(self):
    user_pin=input('enter your pin : ')
    self.pin=user_pin

    user_balance = int(input('enter balance : '))
    self.balance=user_balance

    print('pin created successfully')
    self.menu()

  def change_pin(self):
    old_pin=input('enter old pin : ')

    if old_pin == self.pin:
      #let him change the pin
      new_pin=input('enter new pin : ')
      self.pin=new_pin
      print('pin change successfully! ')
      self.menu()

    else:
      print('nai karne do.')

  def check_balance(self):
    user_pin = input('enter pin : ')
    if user_pin == self.pin:
      print('your balance is ',self.balance)
    else:
      print('nai karne do')

  def withdraw(self):
    user_pin = input('enter pin : ')
    if user_pin == self.pin:
      #allow to withdraw
      amount = int(input('enter the amount you want to withdraw : '))
      if amount <= self.balance:
        self.balance = self.balance - amount
        print('withdrawl successful. balance is ',self.balance)
      else:
        print('insufficient balance')

    else:
      print('nai karne do')
    self.menu()



```

```python
# to make a fraction data type

class Fraction:

  def __init__(self,x,y):    # parametrized constructor ----> constructor which needs input
    self.num = x
    self.den = y

  def __str__(self):      #jb bhi object ko print kraoge to vo __str__ magic method ke pass jaayega aur wha ka code execute ho jaayega
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other):  # ye automatically triger ho jaayega jb 2 objects ke beech mai + ka sign daaloge  # first walla object will be self and 2nd walla other
    print('first fraction is :',self,",",'second fraction is :',other)
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den
    return 'sum of the fractions is : {}/{}'.format(new_num,new_den)

  def __sub__(self, other):    #trigger tb hoga jb 2 objects ko minus kr rhe ho
    print('first fraction is :',self,",",'second fraction is :',other)
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den
    return 'subtraction of the fractions is : {}/{}'.format(new_num,new_den)

  def __mul__(self, other):
    print('first fraction is :',self,",",'second fraction is :',other)
    new_num = self.num*other.num
    new_den = self.den*other.den
    return 'multiplication of the fractions is : {}/{}'.format(new_num,new_den)

  def __truediv__(self, others):
    print('first fraction is :',self,",",'second fraction is :',others)
    new_num = self.num*others.den
    new_den = self.den*others.num
    return 'division of the fractions is : {}/{}'.format(new_num,new_den)

  def convert_to_decimal(self):   #self non magic method
    return self.num/self.den

```

```python
# fr1=Fraction()  ----> will throw error if input not given
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```
